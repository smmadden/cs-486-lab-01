### Dataset: [fifa-19](https://www.kaggle.com/karangadiya/fifa19)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Load and describe dataset:

```python
df = pd.read_csv('data.csv', index_col=0)
df
```

```python
df.describe()
```

```python
df.info()
```

##### Converting the type of the 'Wage' colum to int64:

```python
df['Wage'] = df['Wage'].apply(str.strip, args=['€K']).apply(int)
```

#### Extracted information:

```python
desc = df.describe()
print("Average age of the players:", 
      desc['Age']['mean'])
print("Name of the oldest player: ", 
      df[df['Age'] == desc['Age']['max']]['Name'].iloc[0], " (", desc['Age']['max'], " years old)", sep='')
print("The highest salary: €", 
      desc['Wage']['max'], "K", sep='')
```

### Histogram

```python
print("a)")
plt.hist(df.Age)
plt.title("Player Ages - matplotlib")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
print("b)")
sns.histplot(df.Age)
plt.title("Player Ages - seaborn")
plt.show()
```

These histograms (created with different libraries, depicting the same data) show the frequency of certain values in the given data on player ages. Histogram a) shows that age 20-25 is the most common age for the players, while histogram b) shows more accurately that the most common age is 21 years old. From both of these histograms, we can also conclude that there is not a normal distribution of ages. The majority of players tend to be on the younger side, with fewer and fewer players as the age goes up. This makes sense with our data because as players get older, their physical health, stamina, and speed begin to decline, and injuries are more common. From these histograms, we could hypothesize that ages 20-25 are the most common or even ideal ages for playing soccer/football professionally.

### Correlation

```python
df_corr = df.corr()
df_corr
```

#### Positive and Negative Correlations

```python
df_corr.loc['ShortPassing', 'BallControl']
```

There is a positive correlation between ShortPassing and BallControl in the dataset. This means that as the value for ShortPassing, a rating of the player's skill, goes up, so does BallControl. This makes sense with our data because to be good at short passing to another player in soccer/football, a player needs to be able to control the ball really well.

```python
df_corr.loc['Age', 'Potential']
```

There is a negative correlation between Age and Potential in the dataset. This means that as Age increases, the value for Potential decreases, and vice versa. This makes sense for our dataset because as mentioned earlier in the lab, as a player's age increases, their health, speed, and stamina begin to decline, they have maxed out their skills, and the player's end of career is getting closer and closer. That means that a player's future potential then goes down. On the other hand, the younger players who are maybe more physically healthy and able, who are maybe less skilled, have a lot of potential in the future.

```python
print(df_corr.loc['GKDiving', 'ShortPassing'])
print(df_corr.loc['GKDiving', 'GKHandling'])
```

In this last example, there is a negative correlation between GKDiving and ShortPassing. This makes sense with our dataset because one is a rating of a skill that is specific to goalkeepers (diving) and the other is not. The players that are good at goalkeeping do not regularly practice as the other positions, and vice versa. On the other hand, because GKDiving and GKHandling are both skills attributed to goalkeepers, there is a high positive correlation.

### Further Visualizations

#### Scatter Plot

```python
plt.scatter(x=df.StandingTackle, y=df.SlidingTackle, s=1)
x = range(0, 100)
plt.plot(x, x, color='red', linewidth=0.5)
plt.xlabel('Standing Tackle Rating')
plt.ylabel('Sliding Tackle Rating')
plt.title('Player Tackle Skill')
plt.show()
```

This scatter plot illustrates the very high correlation between the StandingTackle and SlidingTackle skills rating. The red line represents f(x) = x here, and the points in the scatter plot are grouped right on that line. From taking a closer look, it seems like there might be a slight trend towards players being better at standing tackle than sliding tackle, with a few more points on the right side of the line than the left.

#### Line Chart

```python
plt.figure(figsize=(30, 5))
for label, color in [['GKDiving', 'red'], ['GKHandling', 'orange'], ['ShortPassing', 'blue'], ['BallControl', 'green']]:
    plt.plot(df.index, df[label], label=label, color=color)
plt.xlim(0, 100)
plt.legend()
plt.xlabel('Player')
plt.ylabel('Skill Rating')
plt.title('Player Skills')
plt.show()
```

Even without looking at the heatmap from earlier in the lab, this figure shows the positive and negative correlation between the four skills GKDiving, GKHandling, ShortPassing, and BallControl. As shown in the figure above, the highly correlated skills (GKDiving with GKHandling and ShortPassing with BallControl) follow almost the same line. Then, showing the negative correlation between the two pairs, when one pair spikes high, the other drops low.

One thing the heatmap does not show is how even though the goalkeeping skills are negatively correlated to short passing and ball control, these other skills do not drop as low as the goalkeeping skills do for the other players. This makes sense with our data because even the goalkeepers need to be able to control the ball and pass to the other players on their team, while the players that are not goalkeepers have no need to practice or use goalkeeping diving and handling.
